```python
#for implicit  collaborative filtering 
# Simulate 500 users
num_users = 500
user_ids = [f'User{i}' for i in range(1, num_users + 1)]

# Simulate a user-item interaction matrix
np.random.seed(42)
ratings_matrix = pd.DataFrame(
    np.random.randint(1, 6, size=(len(books), num_users)),  # Random ratings between 1-5
    index=books['title'],  # Book titles as rows
    columns=user_ids       # User IDs as columns
)
```

```python
ratings_matrix
```

```python
ratings_matrix = ratings_matrix.applymap(lambda x: x if x >= 1 else 0)
```

```python
ratings_matrix
```

```python
ratings_matrix.index
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_scores=cosine_similarity(ratings_matrix)
```

```python
similarity_scores
```

```python
def recommend(book_name):
    # index fetch
    index=np.where(ratings_matrix.index == book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    data=[]
    for i in similar_items:
        item=[]
        temp_df=books[books['title'] == ratings_matrix.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title'].values))
        item.extend(list(temp_df.drop_duplicates('title')['authors'].values))
        item.extend(list(temp_df.drop_duplicates('title')['thumbnail'].values))
        item.extend(list(temp_df.drop_duplicates('title')['description'].values))
        item.extend(list(temp_df.drop_duplicates('title')['categories'].values))
        
        data.append(item)
        
    return data
```

```python
recommend('Lake Orion')
```

```python
pickle.dump(ratings_matrix,open('ratings_matrix.pkl','wb'))
pickle.dump(similarity_scores,open('similarity_scores.pkl','wb'))
```
